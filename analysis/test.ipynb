```python
import matplotlib.pyplot as plt
import numpy as np

# Define the true mixture proportions
proportions = np.array([0.2, 0.3, 0.5])

# Define the predicted mixture proportions
predictions = np.array([0.15, 0.35, 0.5])

# Calculate the accuracy of the predictions
accuracy = 1 - np.sum(np.abs(predictions - proportions)) / 2

# Plot the true and predicted proportions
plt.bar(['A', 'B', 'C'], proportions, alpha=0.5, label='True proportions')
plt.bar(['A', 'B', 'C'], predictions, alpha=0.5, label='Predicted proportions')
plt.legend()

# Add a title and axis labels
plt.title(f'Accuracy: {accuracy:.2%}')
plt.xlabel('Mixture component')
plt.ylabel('Proportion')

# Display the plot
plt.show()
```

```python
import altair as alt
import pandas as pd
import numpy as np

# Define the true mixture proportions
proportions = np.array([0.2, 0.3, 0.5])

# Define the predicted mixture proportions
predictions = np.array([0.15, 0.35, 0.5])

# Calculate the accuracy of the predictions
accuracy = 1 - np.sum(np.abs(predictions - proportions)) / 2

# Create a dataframe with the true and predicted proportions
df = pd.DataFrame({'component': ['A', 'B', 'C'],
                   'true_proportions': proportions,
                   'predicted_proportions': predictions})

# Define the bar chart
bars = alt.Chart(df).mark_bar().encode(
    x='component:N',
    y=alt.Y('value:Q', title='Proportion'),
    color='variable:N'
).properties(
    title=f'Accuracy: {accuracy:.2%}',
    width=300,
    height=200
)

# Combine the bar chart with a text layer to display the accuracy
accuracy_text = alt.Chart(df).mark_text(
    text=f'Accuracy: {accuracy:.2%}',
    align='center',
    baseline='bottom',
    fontSize=16
).properties(
    width=300,
    height=200
)

# Display the chart
chart = (bars + accuracy_text).configure_axis(
    labelFontSize=14,
    titleFontSize=16
).configure_legend(
    labelFontSize=14
)
chart.show()
```

```python
import altair as alt
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve, auc

# Define the true binary labels and predicted probabilities
y_true = np.array([0, 1, 1, 0, 1, 1, 0, 0, 1, 0])
y_prob = np.array([0.3, 0.7, 0.6, 0.4, 0.8, 0.9, 0.2, 0.1, 0.5, 0.2])

# Compute the false positive rate and true positive rate for different thresholds
fpr, tpr, thresholds = roc_curve(y_true, y_prob)

# Compute the area under the curve (AUC)
auc_score = auc(fpr, tpr)

# Create a dataframe with the ROC curve data
roc_data = pd.DataFrame({'FPR': fpr, 'TPR': tpr, 'Threshold': thresholds})

# Define the ROC curve
roc_curve = alt.Chart(roc_data).mark_line().encode(
    x='FPR:Q',
    y='TPR:Q'
).properties(
    title=f'ROC Curve (AUC={auc_score:.2f})',
    width=400,
    height=400
)

# Add diagonal line
diagonal = alt.Chart(pd.DataFrame({'x': [0, 1], 'y': [0, 1]})).mark_line(color='red').encode(
    x='x:Q',
    y='y:Q'
)

# Display the chart
chart = (roc_curve + diagonal).configure_axis(
    labelFontSize=14,
    titleFontSize=16
)
chart
```
